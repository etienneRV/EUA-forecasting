```python
import pandas as pd
from statsmodels.tsa.stattools import grangercausalitytests

df = pd.read_csv("data/processed/final_feature_matrix.csv", index_col=0, parse_dates=True)

# Create stationary differences for predictors
stationary_vars = df[["d_log_eua", "ttf_gas", "coal_price", "renewable_share", "HDD"]].diff().dropna()

target = "d_log_eua"
for predictor in ["ttf_gas", "coal_price", "renewable_share", "HDD"]:
    print(f"\n=== Does {predictor} Granger-cause EUA returns? ===")
    test_data = stationary_vars[[target, predictor]].dropna()
    # Test up to 5 days of lag
    grangercausalitytests(test_data, maxlag=5, verbose=True)
```
